```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, skew

def plot_distplot(cont_vars, df0, df1, df, figsize=(15, 21)):
    plt.figure(figsize=figsize)
    for i, var in enumerate(cont_vars):
        for j in range(3):
            plt.subplot(7, 3, i * 3 + j + 1)

            if j == 0:
                data = df0[var]
                title = f"Distplot of {var} (Target = 0)"
            elif j == 1:
                data = df1[var]
                title = f"Distplot of {var} (Target = 1)"
            else:
                data = df[var]
                title = f"General Distribution of {var}"

            sns.distplot(data, fit=norm)
            plt.title(title)

            skewness = skew(data.dropna())
            plt.text(0.95, 0.95, f"Skewness: {skewness:.2f}", transform=plt.gca().transAxes, ha='right', va='top')

    plt.tight_layout()
    plt.show()
```

```python
from timeit import default_timer
from time import time
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
model_performance = pd.DataFrame(columns=["model_name", "accuracy", "precision", "recall", "f1_score", "latency"])
def evaluate_model(model, features, labels, model_name=" " ):
    global model_performance
    start_time = time.time()
    predictions = model.predict(features)
    end_time = time.time()
    

    latency = round((end_time - start_time) * 1000, 1)
    accuracy = round(accuracy_score(labels, predictions), 3)
    precision = round(precision_score(labels, predictions, average='binary', zero_division=1), 3)
    recall = round(recall_score(labels, predictions), 3)
    f1 = round(f1_score(labels, predictions, average='binary', zero_division=1), 3)
    

    metrics = {
        "model_name": model_name,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "latency": latency
    }
    model_performance = pd.concat([model_performance, pd.DataFrame([metrics])], ignore_index=True)
    return metrics
    


```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,roc_auc_score,roc_curve
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
def plot_roc_curve(model, feats, labels,title):
    fpr, tpr, thresholds = roc_curve(labels, model.predict_proba(feats)[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(10,8))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

def detect_outliers(data, contamination=0.01, random_state=42):
   
    cat_cols = data.select_dtypes(include='bool').columns
    cont_cols = data.select_dtypes(exclude='bool').columns

    scaler = StandardScaler()
    data_scaled = data.copy()
    data_scaled[cont_cols] = scaler.fit_transform(data_scaled[cont_cols])

    model = IsolationForest(contamination=contamination, random_state=random_state)
    model.fit(data_scaled)
    scores = model.score_samples(data_scaled)
    threshold = np.percentile(scores, 100 * contamination)
    return scores < threshold

```
